```python
import seaborn as sns
import matplotlib.pyplot as plt


amazonia_bolsa_familia_pct = 50  
rs_bolsa_familia_pct = 17.5    


amazonia_nao_beneficiaria_pct = 100 - amazonia_bolsa_familia_pct
rs_nao_beneficiaria_pct = 100 - rs_bolsa_familia_pct

# Tema do Seaborn
sns.set_theme(style='dark', palette='bright', context='notebook')

# Criando a figura e os eixos para os grficos
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# Cores Paleta Seaborn
cores = sns.color_palette("bright", 2)
largura = 0.4

# Alterando cor de fundo e adicionando linhas horizontais para ambos os grficos
for ax in axs:
    ax.set_facecolor('LightGray')  
    ax.grid(axis='y', linestyle='-', alpha=1)  

# Grfico para Amaznia (Regio Norte) - barras
categoria_amazonia = ["Beneficiria", "No Beneficiria"]
porcentagens_amazonia = [amazonia_bolsa_familia_pct, amazonia_nao_beneficiaria_pct]

sns.barplot(x=categoria_amazonia, y=porcentagens_amazonia, ax=axs[0],
            width=largura, palette=[cores[0], cores[1]])
axs[0].set_ylim(0, 100) 
axs[0].set_ylabel('Porcentagem (%)')
axs[0].set_title('Amaznia (Regio Norte)', fontsize=15)

# Adicionar valores nas barras
for i, v in enumerate(porcentagens_amazonia):
    axs[0].text(i, v + 2, f"{v:.1f}%", ha='center', va='bottom', fontsize=10)


# Grfico para Rio Grande do Sul - barras
categoria_rs = ["Beneficiria", "No Beneficiria"]
porcentagens_rs = [rs_bolsa_familia_pct, rs_nao_beneficiaria_pct]

sns.barplot(x=categoria_rs, y=porcentagens_rs, ax=axs[1],
            width=largura, palette=[cores[0], cores[1]])

axs[1].set_ylim(0, 100)  
axs[1].set_ylabel('Porcentagem (%)')
axs[1].set_title('Rio Grande do Sul', fontsize=15)
# Adicionar valores nas barras
for i, v in enumerate(porcentagens_rs):
    axs[1].text(i, v + 2, f"{v:.1f}%", ha='center', va='bottom', fontsize=10)

# ajustes finais ---------------------------------------------------------------------
fig.suptitle('''Comparativo de Famlias Beneficirias do Bolsa Famlia
Amaznia (Regio Norte) vs. Rio Grande do Sul''', fontsize=18)

# Ajustar layout com espaamento
plt.subplots_adjust(wspace=0.4)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
